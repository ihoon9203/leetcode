# 2. Add Two Numbers

## <span style="color:orange">Medium</span>


You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

 

Example 1:


Input: l1 = [2,4,3], l2 = [5,6,4]
Output: [7,0,8]
Explanation: 342 + 465 = 807.
Example 2:

Input: l1 = [0], l2 = [0]
Output: [0]
Example 3:

Input: l1 = [9,9,9,9,9,9,9], l2 = [9,9,9,9]
Output: [8,9,9,9,0,0,0,1]
 

### Constraints:

The number of nodes in each linked list is in the range [1, 100].
0 <= Node.val <= 9
It is guaranteed that the list represents a number that does not have leading zeros.

````python
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        head = ListNode()
        ptrNode = head
        carry = 0
        while l1 or l2 or carry:
            v1 = l1.val if l1 else 0
            v2 = l2.val if l2 else 0
            # assign value
            newVal = v1 + v2 + carry
            carry = newVal // 10
            newVal = newVal % 10
            ptrNode.next = ListNode(newVal)

            # moving ptr\
            ptrNode = ptrNode.next
            l1 = l1.next if l1 else None
            l2 = l2.next if l2 else None
        return head.next
            
````

## 해설

이건 사실 덧셈의 원리를 LinkedList로 풀어쓴 문제다.

l1과 l2의 첫째자리 수를 더하고 10 이상이면 다음에 1을 더하게 하는 원리임.

``` Python
v1 = l1.val if l1 else 0
v2 = l2.val if l2 else 0
```
만약 맞지 않는 digit 예를 들어 90999, 99라면 l2는 None 일텐데 그 대신 0을 넣어줌

``` Python
newVal = v1 + v2 + carry
carry = newVal // 10
newVal = newVal % 10
ptrNode.next = ListNode(newVal)
```

새로운 자릿 수의 값은 v1 + v2 + carry
l1과 l2의 각 자릿수 더하기 carry (만약 이전 자리값의 수의 합이 10 이상이였을 경우)

``` Python
while l1 or l2 or carry:
```

만약 
``` Python
while l1 or l2:
```
라고 했으면 대부분 가능했겠지만 만약 99 + 9 같은 케이스라면 최종 자릿수가 l1 과  l2 보다 더 큰 108이 된다. 그래서 l1과 l2가 모두 없더라고 carry가 있으면 이를 고려해야함
